```python
from astropy.io.votable import parse_single_table
import numpy as np
import os
import glob
from astropy.io import fits
#from tqdm import tdqm
```

```python
f = open("wgets.txt", "r")
specs = np.array(f.readlines())
```

```python
for i in range(len(specs)):
    specs[i] = specs[i].strip('\n')
```

```python
n_points = len(specs)

indx = np.linspace(0, len(specs)-1, n_points, dtype = int)
download = specs[indx]
```
