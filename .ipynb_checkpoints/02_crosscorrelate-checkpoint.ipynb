```python
from scipy.interpolate import RegularGridInterpolator
import numpy as np

T_max = 7100
T_min = 2000
T_step = 100
teffs = np.arange(T_min, T_max, T_step)

logg_max = 5.5
logg_min = 2.5
logg_step = 0.5
loggs = np.arange(logg_min, logg_max, logg_step)

rv_max = 250
rv_min = -250
rv_step = 25
rvs = np.arange(rv_min, rv_max, rv_step)

values = np.zeros((len(teffs), len(loggs), len(rvs), 6000))

wl_grid = np.linspace(4000, 10000, 6000)

for i in range(len(teffs)):
    for j in range(len(loggs)):
        for k in range(len(rvs)):
            wl = grid[teffs[i]][loggs[j]][rvs[k]].T[0]
            fl = grid[teffs[i]][loggs[j]][rvs[k]].T[1]
            
            #print(wl)
            #break
            
            fl_new = np.interp(wl_grid, wl, fl)
            
            grid[teffs[i]][loggs[j]][rvs[k]] = np.array([fl_new])
            values[i,j,k] = np.array([fl_new])
```

```python
interp_spec = RegularGridInterpolator((teffs, loggs, rvs),values)
```

```python
import lmfit

def resample(wl, interp, params):
    teff, logg, rv = params[:3]
    
    return(np.interp(wl, wl_grid, interp_spec(params)))

def get_spectrum(params, wl = wl_grid):
    vals = (params['teff'], params['logg'], params['rv'])
    
    return(np.interp(wl, wl_grid, interp_spec(vals)))

def residual(params, wl = wl_grid, spec = None, ivar = None):
    model_spec = get_spectrum(params, wl)
    
    #print(spec)
    
    if ivar is not None:
        chisquare = sum(((model_spec - spec) * ivar)**2)
    else:
        chisquare = sum(((model_spec - spec))**2)
        
    #return 1/2*np.sum(((spec - model_spec)** 2 / (ivar)**-1 + np.log( (ivar)**-1 )))
    
    return chisquare
```

```python
def get_parameters(spec, wl = wl_grid, ivar = None, p0 = [5000, 4, 0]):          
    params = lmfit.Parameters()
    
    params.add('teff', value = p0[0], min = 2000, max = 7000, vary = True)
    params.add('logg', value = p0[1], min=2.5, max=5, vary=True)
    params.add('rv', value = p0[2], min = -250, max = 225, vary = True)
        
    result = lmfit.minimize(residual, params, kws = dict(wl = wl_grid, spec = spec, ivar = ivar), method = 'emcee', nan_policy='omit', burn=300, 
                    steps=6000, nwalkers = 1000, thin=20, is_weighted=True, float_behavior = 'chi2')
            
    return result
```

```python
pts = (6734, 3.1, 10)

fl_test = interp_spec(pts)
ivar_test = np.ones(len(fl_test))*10

results = get_parameters(fl_test, ivar = ivar_test, p0 = [3000, 4, 1])
```

```python
import corner

emcee_plot = corner.corner(results.flatchain, labels=results.var_names,
                           truths=list(results.params.valuesdict().values()))
```
