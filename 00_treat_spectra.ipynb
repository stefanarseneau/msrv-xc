```python
import glob
import os

import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
from astropy.convolution import convolve_fft
from astropy.convolution import Gaussian1DKernel

def smooth(wl, fl, R, w_sample = 1):
    # Generate logarithmic wavelength grid for smoothing
    w_logmin = np.log10(np.nanmin(wl))
    w_logmax = np.log10(np.nanmax(wl))
    n_w = np.size(wl)*w_sample
    w_log = np.logspace(w_logmin, w_logmax, num=n_w)
    
    # Find stddev of Gaussian kernel for smoothing
    R_grid = (w_log[1:-1]+w_log[0:-2])/(w_log[1:-1]-w_log[0:-2])/2
    sigma = np.median(R_grid)/R
    if sigma < 1:
        sigma = 1
        
    # Interpolate on logarithmic grid
    f_log = np.interp(w_log, wl, fl)
    
    # Smooth convolving with Gaussian kernel
    gauss = Gaussian1DKernel(stddev=sigma)
    f_conv = convolve_fft(f_log, gauss)

    # Interpolate back on original wavelength grid
    f_sm = np.interp(wl, w_log, f_conv)
    
    return wl, f_sm

from numpy.polynomial.polynomial import polyfit,polyval

def pseudonormalize(fl, win = 50):
    cont_fl = np.repeat(np.nan, len(fl))
    
    for i in range(len(fl)):
        if (i - (win / 2)) < 0:
            fl_win = fl[0:(i + (win // 2))]
        else:
            fl_win = fl[(i - (win // 2)) : (i + (win // 2))]
        
        cont_fl[i] = np.percentile(fl_win, 90)
        
    return fl / cont_fl
```

```python
files = glob.glob('phoenix/*.fits.gz')
print(files)
```

```python
path = 'phoenix/convolved/' + str(files[0].split('/')[-1].split('.spec')[0]) + '.fits'
print(path)
```

```python
from astropy.table import Table

i = 0
j = len(files)
for file in files:
    print('Progress: {} / {}'.format(i, j))
    spec = Table.read(file)
        
    new_wl, new_fl = smooth(spec['Wavelength'], spec['Flux'], 2000)
    newer_fl = pseudonormalize(new_fl, 100)
    
    spec['smooth_fl'] = new_fl
    spec['norm_fl'] = newer_fl
    
    path = 'phoenix/convolved/' + str(file.split('/')[-1].split('.spec')[0]) + '.fits'
    spec.write(path, overwrite = True)
    
    i += 1
```
